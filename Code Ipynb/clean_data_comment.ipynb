```python
import pandas as pd
```

```python
file_path = "comments_data_ncds.csv"
df = pd.read_csv(file_path)
```

```python
print(df.info())
```

```python
# Hiển thị 5 dòng đầu tiên
print("\n5 dòng đầu tiên:")
print(df.head())
```

```python
# Xoá các cột hoàn toàn không có dữ liệu và không cần thiết
columns_to_drop = ["id", "thank_count"]
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
# Hiển thị 5 dòng đầu tiên
print("\n5 dòng đầu tiên:")
print(df_cleaned.head())
```

```python
# Chuyển các cột có kiểu 'object' thành 'string'
for column in df_cleaned.select_dtypes(include=['object']).columns:
    df_cleaned[column] = df_cleaned[column].astype('string')
print(df_cleaned.info())
```

```python
print(df_cleaned.info())
```

```python
df_cleaned.to_csv("cleaned_comment_data.csv", index=False)
```
